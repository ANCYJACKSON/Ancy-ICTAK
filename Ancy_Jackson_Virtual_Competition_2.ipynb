```python
bins=[0,2500,4000,6000,81000]
group=['Low','Average','High', 'Very high']
trn_data['Income_bin']=pd.cut(trn_data['ApplicantIncome'],bins,labels=group)
trn_data.head()
```

```python
Income_bin=pd.crosstab(trn_data['Income_bin'],trn_data['Loan_Status'])
Income_bin.div(Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('ApplicantIncome')
P = plt.ylabel('Percentage')
```

It can be inferred that Applicant’s income does not affect the chances of loan approval which contradicts our hypothesis in which we assumed that if the applicant’s income is high the chances of loan approval will also be high.

```python
bins=[0,1000,3000,42000]
group=['Low','Average','High']
trn_data['Coapplicant_Income_bin']=pd.cut(trn_data['CoapplicantIncome'],bins,labels=group)
trn_data.head()
```

```python
Coapplicant_Income_bin=pd.crosstab(trn_data['Coapplicant_Income_bin'],trn_data['Loan_Status'])
Coapplicant_Income_bin.div(Coapplicant_Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('CoapplicantIncome')
P = plt.ylabel('Percentage')
```

It shows that if co-applicants income is less the chances of loan approval are high. But this does not look right. So we are making a new variable in which we will combine the applicant’s and co applicants’ income to visualize the combined effect of income on loan approval.

```python
trn_data['Total_Income']=trn_data['ApplicantIncome']+trn_data['CoapplicantIncome']
bins=[0,2500,4000,6000,81000]
group=['Low','Average','High', 'Very high']
trn_data['Total_Income_bin']=pd.cut(trn_data['Total_Income'],bins,labels=group)
trn_data.head()
```

```python
Total_Income_bin=pd.crosstab(trn_data['Total_Income_bin'],trn_data['Loan_Status'])
Total_Income_bin.div(Total_Income_bin.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True)
plt.xlabel('Total_Income')
P = plt.ylabel('Percentage')
```

We can see that Proportion of loans getting approved for applicants having low Total_Income is very less as compared to that of applicants with Average, High, and Very High Income.

## Checking Missing values - Train dataframe

```python
trn_data.isna().sum()
```

```python
plt.hist(trn_data['LoanAmount'])
```

This is positively skewed data, so using the ‘fillna’ method for imputing the column 'LoanAmount' with the median value.

```python
trn_data['LoanAmount']= trn_data['LoanAmount'].fillna(trn_data['LoanAmount'].median())
```

```python
plt.hist(trn_data['Loan_Amount_Term'])
```

```python
trn_data['Loan_Amount_Term'].value_counts()
```

It can be seen that in loan amount term variable, the value of 360 is repeating the most. So we will replace the missing values in this variable using the mode of this variable.

This is negatively skewed data, so using the ‘fillna’ method for imputing the column 'Loan_Amount_Term' with the mode value.

```python
trn_data['Loan_Amount_Term'].fillna(trn_data['Loan_Amount_Term'].mode().values[0],inplace=True)
```

```python
plt.hist(trn_data['Credit_History'])
```

This is negatively skewed data, so using the ‘fillna’ method for imputing the column 'Credit_History'

For numerical variables: imputation using mean or median

For categorical variables: imputation using mode

```python
trn_data['Credit_History'].fillna(trn_data['Credit_History'].mode().values[0],inplace=True)
```

```python
#  Fill all the missing values with mod for the object fields
trn_data['Gender'].fillna(trn_data['Gender'].mode().values[0],inplace=True)
trn_data['Married'].fillna(trn_data['Married'].mode().values[0],inplace=True)
trn_data['Dependents'].fillna(trn_data['Dependents'].mode().values[0],inplace=True)
trn_data['Self_Employed'].fillna(trn_data['Self_Employed'].mode().values[0],inplace=True)
```

### Handling outliers for 'LoanAmount' field

```python
trn_data['LoanAmount'].skew()
```

Inference:

Outliers often has a significant effect on the mean and standard deviation and hence affecting the distribution. Due to these outliers bulk of the data in the LoanAmount is highly positively skewed.

So we must remove outliers from our data sets. So we are adopting the log transformation, which does not affect the smaller values much, but reduces the larger values. So, we get a distribution similar to normal distribution.

```python
trn_data['LoanAmount'].dtypes
```

```python
trn_data['LoanAmount_log' ] = np.log(trn_data['LoanAmount'])
trn_data['LoanAmount_log'].hist (bins=20)
```

## Inference :

## Now the distribution looks much closer to a normal distribution and effect of extreme values has been significantly subsided.
