A company started to invest in digital marketing as a new way of their product promotions. For that they collected data and decided to carry out a study on it.

  ● The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

  ● The company needs to check whether there is any dependency between the features “Region” and “Manager”.

Help the company to carry out their study with the help of data provided.

```python
daf = pd.read_csv('/content/Sales_add.csv')
```

```python
daf
```

```python
daf.info()
```

```python
# Question 1
# The company wishes to clarify whether there is any increase in sales after stepping into digital marketing
#       H0, Null Hypothesis : No improvement in sales after stepping into digital marketing.
#       H1, Alternate Hypothesis : There is an improvement in sales after stepping into digital marketing.
```

# TWo Sample TTest (Sample size < 30)

```python
from scipy.stats import ttest_ind
```

```python
bef_sales = daf['Sales_before_digital_add(in $)']
aft_sales = daf['Sales_After_digital_add(in $)']
```

```python
t_stat,p_value = ttest_ind(bef_sales,aft_sales)
```

```python
t_stat
```

```python
p_value
```

```python
if p_value < 0.05:
    print('Reject Null Hypothesis')
else:
    print('Accept Null Hypothesis')
```

# Inference : Rejecting the Null Hypothesis, so there is an improvement in sales after implementing digital marketing.

```python
# Question 2
# The company needs to check whether there is any dependency between the features “Region” and “Manager”.
#       H0, Null Hypothesis : No dependancy between “Region” and “Manager”.
#       H1, Alternate Hypothesis : There is a dependancy between “Region” and “Manager”.
```

# CHI SQUARE test (Test Of Independance) - compare two categorical variables(Region, Manager) in a contingency table

```python
from scipy.stats import chi2_contingency
```

```python
mgr_reg = pd.crosstab(daf['Region'], daf['Manager'])
mgr_reg
```

```python
chi_stat,p_value,dof,expected = chi2_contingency(mgr_reg)
```

```python
print('Statistic:', chi_stat)
print('p_value:', p_value)
print('dof:', dof)
print('expected freq:', expected)
```

```python
if p_value < 0.05:
    print('Reject Null Hypothesis')
else:
    print('Accept Null Hypothesis')
```

# Inference : Accepting the Null Hypothesis, so there is no dependancy between “Region” and “Manager” .
